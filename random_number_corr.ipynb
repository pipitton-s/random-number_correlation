```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Linear congruential generator, x =seed

x =1
n = 32
rand = []
# print(x)
for i in range(n):
    x_1 = (5*x+7)%n
    print(x_1)
    rand.append(x_1)
    x = x_1
```

```python
np.polyfit(np.arange(n),rand,1)
fit = np.array([-0.25513196*n+19.45454545 for n in range(n)])
```

```python
plt.scatter(np.arange(n),rand)
plt.plot(fit,'red')
plt.title('Random number output')
plt.legend(['number',r'corr $\approx$ -0.25'])
plt.show()
```

```python
fig,ax = plt.subplots(16,figsize=(10,80))
# ax[0,0].plot([1,2,3,4],[1,2,3,4])
for i in range(16):
        l = 1+i
        corr = []
        for N in range(n-l):
            corr.append(abs(rand[N]-rand[N+l]))

        # ax[i,j].scatter(i,j)
        ax[i].scatter(np.arange(len(corr)),corr)
        ax[i].set_title('correlation l=' + str(l))
# plt.show()
```
